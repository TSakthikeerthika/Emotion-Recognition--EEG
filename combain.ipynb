# Load datasets :

```python
import os
import scipy.io

# List of file names you want to load
file_names_to_load = ['deap_1.mat', 'deap_2.mat', 'deap_3.mat', 'deap_4.mat', 'deap_5.mat', 'deap_6.mat', 'deap_7.mat', 'deap_8.mat', 'deap_9.mat', 'deap_10.mat', 'deap_11.mat', 'deap_12.mat', 'deap_13.mat', 'deap_14.mat', 'deap_15.mat', 'deap_16.mat', 'deap_17.mat', 'deap_18.mat', 'deap_19.mat', 'deap_20.mat', 'deap_21.mat', 'deap_22.mat', 'deap_23.mat', 'deap_24.mat', 'deap_25.mat', 'deap_26.mat', 'deap_27.mat', 'deap_28.mat', 'deap_29.mat', 'deap_30.mat', 'deap_31.mat','deap_32.mat']

# Specify the directory path where .mat files are located
directory_path = '/content/drive/MyDrive/dataset folder'
```

# combain datasets :

```python
# Create an empty NumPy array to store the data
all_data = np.empty((0, 416), dtype=np.float32)

# Load each specified .mat file
for filename in file_names_to_load:
    file_path = os.path.join(directory_path, filename)
    if os.path.isfile(file_path):
        mat_data = scipy.io.loadmat(file_path)

        # Access specific data within the loaded file
        if 'data' in mat_data:
            data = mat_data['data']

            # Append the data to the 'all_data' array
            all_data = np.vstack((all_data, data))
        else:
            print(f"'data' key not found in file '{filename}'")

        # Perform your desired operations on the data
        # ...

        # Example: Print the file name and data shape
        print(f"Loaded file: {filename}")
        print(f"Data shape: {data.shape}")
        print()
    else:
        print(f"File '{filename}' does not exist in the directory.")

# Verify the shape of the combined data
print(f"All Data shape: {all_data.shape}")
```

# save combain (all 32 person) data :

```python
# Save the combined data as a .mat file
save_file_path = '/content/drive/MyDrive/dataset folder/all_32_deap.mat'

# Prepare the data dictionary to be saved
data_dict = {'all_32_data-deap': all_data}

# Use the scipy.io.savemat() function to save the data
scipy.io.savemat(save_file_path, data_dict)

print(f"Data saved as '{save_file_path}'.")
```

# load labels , save :

```python
 # Print the file names
file_names_load = ['s{:02d}.mat'.format(i) for i in range(1, 33)]

for filename in file_names_to_load:
    print(filename)

# List to store the labels from all files
all_labels = []

# Load each specified .mat file
for filename in file_names_load:
    file_path = os.path.join(directory_path, filename)
    if os.path.isfile(file_path):
        mat_data = scipy.io.loadmat(file_path)

        # Access the 'labels' key
        if 'labels' in mat_data:
            labels = mat_data['labels']
            all_labels.append(labels)
        else:
            print(f"'labels' key not found in file '{filename}'")
    else:
        print(f"File '{filename}' does not exist in the directory.")

# Combine the labels into a single numpy array
all_32_labels = np.vstack(all_labels)

# Verify the shape of the combined labels
print(f"All Labels shape: {all_32_labels.shape}")
```
